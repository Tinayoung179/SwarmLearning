```python
from tensorflow.keras.preprocessing.image import load_img, img_to_array
def load_images(file_paths, size=(28, 28)):
    images = []
    for path in file_paths:
        # sử dụng keras.preprocessing.image.load_img để đọc ảnh
        image = load_img(path, target_size=size, color_mode='grayscale')
        # chuyển ảnh về dạng ma trận và thêm vào list
        images.append(img_to_array(image))
    # chuyển list thành một ma trận numpy
    images = np.array(images, dtype='float32') / 255.0
    return images
```
