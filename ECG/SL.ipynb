```python
import pandas as pd
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.python.keras.layers.core import Dense
from tensorflow.keras.models import Sequential
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from tensorflow.keras.utils import to_categorical
```

# Load dữ liệu test

```python
data = pd.read_csv('/content/drive/MyDrive/Khoa_luan/ECG_flip/ECG_heatbeat_20/app-data/Test.csv')
y_test = data.iloc[: , -1]
X_test = data.iloc[: , :-1]
num_records = X_test.shape[0]
X_test = tf.constant(X_test.values[:,0:187])
X_test = tf.reshape(X_test, (num_records, 187, 1))

y_test = tf.one_hot(y_test.values, 2)
y_test = tf.reshape(y_test, (num_records, 2))
```

# Dữ liệu bình thường

```python
import tensorflow as tf
from tensorflow import keras
# Load the model from the .pb file
model = keras.models.load_model('/content/drive/MyDrive/SwarmLearn/Cardio/User1/Model_mitbih.pb')
```

```python
from keras.optimizers import SGD
sgd = tf.keras.optimizers.SGD(learning_rate=0.01)
model.compile(loss = 'binary_crossentropy', optimizer=sgd, metrics=['accuracy'])
```

```python
y_pred = model.predict(X_test)
y_pred= np.argmax(y_pred, axis=1)


# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# Calculate precision
precision = precision_score(y_test, y_pred)
print("Precision:", precision)

# Calculate recall
recall = recall_score(y_test, y_pred)
print("Recall:", recall)

# Calculate F1-score
f1score = f1_score(y_test, y_pred)
print("F1-score:", f1score)
```
